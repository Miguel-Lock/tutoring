```python
# Variables, types, printing, and basic I/O
name = "Nano"
year = 2018
weight_m = 27.2

print("Hello, ", name)
print(f"Welcome to {year}. Your weight is {weight_m} pounds.")

print(type(name), type(year), type(weight_m))

# Try converting types
year_str = str(year)
print("Converted year to string:", year_str, type(year_str))

# Practice: Make your own variables and print them using an f-string

intelligence_map = {'Seth': 70, 'Able': 80, 'Cain': 2,}
smart_people = []

for person, intelligence in intelligence_map.items():
  if intelligence > 50:
    smart_people.append(person)

print(type(intelligence_map))
print(type(smart_people))
print(f"{', '.join(smart_people)} have the intelligence of more than 50!")
```

```python
# Importing modules (math)
# Typically, you want to do this at the start of the notebook
# it works mid cells too, just be mindful of the order in which you run cells.
import math

r = 2.5
circle_area = math.pi * (r**2)
print(f"For r={r}, area={circle_area:.3f}")

# math functions
print(math.sin(math.pi/2), math.cos(0), math.sqrt(16))

# Practice: Use math functions to compute
print("factorial of 3:", math.factorial(3))
print('fmod(17, 7):', math.fmod(17, 7))
```

```python
import math

# Lists and basic operations
nums = [2, 4, 6, 8, 10]
print("nums:", nums)
print("index 0:", nums[0], "slice 1:4:", nums[1:4])
nums.append(12)
print("after append:", nums)

# List comprehension: squares of 0..9
squares = [i*i for i in range(10)]
print("squares:", squares)

# Practice: Make a list of 5 floats and compute their average
meFloats = [3.1415, 1.23, 4.56, 7.89, 10.1112]
count = 0
sum = 0
for item in meFloats:
  sum += item
  count += 1
print(f"Average of the numbers: {', '.join([str(float) for float in meFloats])}")
print(sum / count)
```

```python
# Matplotlib basics
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.figure()
plt.plot(x, y, marker='o', linestyle='-')  # no explicit color
plt.title("Quadratic Growth")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.legend(["data"])
plt.show()

# TODO: Make a scatter plot instead of a line plot
plt.scatter(np.array(x), np.array(y))
plt.title("Quadratic Growth Scattered")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
plt.show()
```

```python
import math

# f-strings and LaTeX in labels
n = 2
label = f"$y=\\sin({n}x)$"

xs = [i * 0.1 for i in range(0, 63)]  # ~0..6.2
ys = [math.sin(n * val) for val in xs]

plt.figure()
plt.plot(xs, ys)  # no explicit color
plt.title("Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel(label)
plt.legend([label])
plt.show()

# Practice: Try cosine and change the title accordingly
label = f"$y=\\cos({n}x)$"
xc = [i * 0.1 for i in range(0, 63)]
yc = [math.cos(n * val) for val in xc]

plt.figure()
plt.plot(xc, yc)  # no explicit color
plt.title("Cosine Wave")
plt.xlabel("x (radians)")
plt.ylabel(label)
plt.legend([label])
plt.show()
```
